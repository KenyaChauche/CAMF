```python
import pandas as pd
import numpy as np
```

```python
units = pd.read_excel("../data/CAMF Data Pull 7 rlmain.xlsx")
incidents = pd.read_excel("../data/CAMF Data Pull 6 frmain.xlsx")
```

```python
units = units[units['callid'].notnull()].copy().reset_index()

units.head(20)
```

```python
incidents.head()
```

```python
units['callid'] = [unit.lstrip() for unit in units['callid']]

incidents.drop(1878, inplace = True)
incidents['callid'] = [incident.lstrip() for incident in incidents['callid']]
```

```python
combined = {
    'callid': [], 
    'apparatus': [], 
    'zone': [],
    'address': [], 
    'alarm_date': [], 
    'en_route_date': [], 
    'arrival_date': [], 
    'hospital_arrival_date': [], 
    'hospital_departure_date': [], 
    'clear_date': [], 
    'call_type': []
}

callids = units['callid'].unique()
```

```python
for call in callids: 
    apparatus = list(units[units['callid'] == call]['desc'].unique())
    num_units = len(apparatus)
#     print(call)
    for n in range(0, num_units): 
#         print(apparatus[n])
        combined['callid'].append(call)
        combined['apparatus'].append(apparatus[n])
        combined['zone'].append(units[units['callid'] == call]['zone'].unique()[0])
        try: 
            combined['address'].append(incidents[incidents['callid'] == call]['address'].unique()[0])
        except: 
            combined['address'].append(None)
        
        try: 
            combined['alarm_date'].append(units[(units['callid'] == call) & (units['desc'] == apparatus[n]) & 
                                                (units['logdate'] == 'PAGED')]['unit'].unique()[0])
        except: 
            combined['alarm_date'].append(None)
        
        try: 
            combined['en_route_date'].append(units[(units['callid'] == call) & (units['desc'] == apparatus[n]) & 
                                                   (units['logdate'] == 'ENRT')]['unit'].unique()[0])
        except: 
            combined['en_route_date'].append(None)
        
        try: 
            combined['arrival_date'].append(units[(units['callid'] == call) & (units['desc'] == apparatus[n]) & 
                                                  (units['logdate'] == 'ARRVD')]['unit'].unique()[0])
        except: 
            combined['arrival_date'].append(None)
        
        try: 
            combined['hospital_arrival_date'].append(units[(units['callid'] == call) & (units['desc'] == apparatus[n]) & 
                                                           (units['logdate'] == 'HOSP')]['unit'].unique()[0])
            
        except: 
            combined['hospital_arrival_date'].append(None)
            
        try:     
            combined['hospital_departure_date'].append(units[(units['callid'] == call) & (units['desc'] == apparatus[n]) & 
                                                             (units['logdate'] == 'ENRT')]['unit'].unique()[0])
            
        except:
            combined['hospital_departure_date'].append(None)
        
        try: 
            combined['clear_date'].append(units[(units['callid'] == call) & (units['desc'] == apparatus[n]) & 
                                                (units['logdate'] == 'CMPLT')]['unit'].unique()[0])
        except: 
            combined['clear_date'].append(None)
        
        try: 
            combined['call_type'].append(incidents[incidents['callid'] == call]['nature'].unique()[0])
        except: 
            combined['call_type'].append(None)
            
print('Done')
```

```python
combined_df = pd.DataFrame(combined)
```

```python
combined_df.to_excel('../data/CAMF combined 2019.xlsx')
```
