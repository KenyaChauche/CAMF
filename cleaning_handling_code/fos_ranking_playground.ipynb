```python
import pandas as pd

df = pd.read_excel("../data/ESO 2020Q1.xls", header = 2)
```

```python
df.head()
```

```python
list(df[df['Incident Number'] == '20-0371'].sort_values(by = ['Arrival Date'])['Apparatus Name'].values)
```

```python
def arrival_rank(incident_num, apparatus): 
    try: 
        ordered_list = list(df[df['Incident Number'] == 
                               incident_num].sort_values(by = ['Arrival Date'])['Apparatus Name'].values)
        return ordered_list.index(apparatus) + 1
    except: 
        return None
```

```python
arrival_rank('20-0371', 'E14')
```

```python
df['Rank of Arrival'] = [arrival_rank(df.loc[i, 'Incident Number'], df.loc[i, 'Apparatus Name']) for i in df.index]
```
